```python
import os
import pandas as pd
import yaml

def yaml_to_csv(input_folder,output_folder):
    os.makedirs(output_folder,exist_ok=True)
    
    for month in os.listdir(input_folder):
        month_path=os.path.join(input_folder,month)
        if not os.path.isdir(month_path):
            continue

        for file in os.listdir(month_path):
            file_path=os.path.join(month_path,file)
            if not file.endswith(".yaml"):
                continue

            print(f"file processsing:{file_path}")

            try:
                with open(file_path,"r")as yaml_file:
                    data=yaml.safe_load(yaml_file)
                    df=pd.DataFrame(data)
                    symbol=file.split(".")[0]
                    
                    if "date" in df.columns:
                        df['date']=pd.to_datetime(df['date'],format="%Y-%m-%d %H:%M:%S") .dt.date
                    
                    output_path=os.path.join(output_folder,f"{symbol}.csv")
                    df.to_csv(output_path,index=False)
                    print(f"saved csv:{output_path}")

            except Exception as e:
                print("error occur{file_path}:{e}")

yaml_to_csv("D:/projects/stock/data", "D:/projects/stock/outputcsv")               


```

```python
import pandas as pd
from pathlib import Path  

file_path=Path("D:/projects/stock/outputr/combined.csv")
output_folder=Path('D:/projects/stock/outputsi')
output_folder.mkdir(exist_ok=True)

df=pd.read_csv(file_path)

if "Ticker" not in df.columns:
    raise ValueError("ticker column not found")

for ticker, data in df.groupby("Ticker"):
    output_path=output_folder.joinpath(f"{ticker}.csv")
    data.to_csv(output_path,index=False)
    print(f"saved csv:{output_path}")

 
```
